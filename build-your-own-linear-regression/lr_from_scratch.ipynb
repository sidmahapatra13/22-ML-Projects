<h1>Objective</h1>
<p>The aim of this project is to build a <b>Linear Regression model</b> from scratch using Python, without relying on machine learning libraries such as Scikit-learn. The project focuses on understanding the mathematical foundation of linear regression by implementing the cost function, gradient descent optimization, and parameter updates manually. The central goal is to develop a model that learns the relationship between input features and a continuous target variable while gaining a deeper understanding of how regression algorithms work internally.</p>
<hr>

```python
class MyLR:

    def __init__(self):
        self.m = None
        self.b = None

    def fit(self, X_train, y_train):

        num = 0
        den = 0

        for i in range(X_train.shape[0]):
            num += (X_train[i] - X_train.mean()) * (y_train[i] - y_train.mean())
            den += (X_train[i] - X_train.mean()) ** 2

        self.m = num / den
        self.b = y_train.mean() - self.m * X_train.mean()

        print("Slope:", self.m)
        print("Intercept:", self.b)

    def predict(self, X_test):
        return self.m * X_test + self.b
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,0].values
y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
```

```python
lr = MyLR()
lr.fit(X_train, y_train)
```

<h1>Evaluation Metrics</h1>
<p>Predictions from the single test point are replaced with metrics computed over the <b>entire test set</b>: MAE, MSE, RMSE, and R&sup2;.</p>

```python
y_pred = lr.predict(X_test)

mae = np.mean(np.abs(y_test - y_pred))
mse = np.mean((y_test - y_pred) ** 2)
rmse = np.sqrt(mse)

ss_res = np.sum((y_test - y_pred) ** 2)
ss_tot = np.sum((y_test - y_test.mean()) ** 2)
r2 = 1 - (ss_res / ss_tot)

print("MAE:", mae)
print("MSE:", mse)
print("RMSE:", rmse)
print("R2 Score:", r2)
```

<h1>Visualization</h1>
<p>Scatter plot of the training data with the fitted regression line, plus the test points overlaid for a visual sanity check.</p>

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
plt.scatter(X_train, y_train, color='blue', label='Train data')
plt.scatter(X_test, y_test, color='green', label='Test data')

x_line = np.linspace(X.min(), X.max(), 100)
y_line = lr.predict(x_line)
plt.plot(x_line, y_line, color='red', label='Fitted line (MyLR)')

plt.xlabel(df.columns[0])
plt.ylabel(df.columns[1])
plt.title('Linear Regression Fit')
plt.legend()
plt.show()
```

<h1>Sanity Check: Compare with Scikit-learn</h1>
<p>Fitting scikit-learn's <code>LinearRegression</code> on the same split to confirm <code>MyLR</code> learns the same slope and intercept.</p>

```python
from sklearn.linear_model import LinearRegression

sk_lr = LinearRegression()
sk_lr.fit(X_train.reshape(-1, 1), y_train)

print("Scikit-learn slope:", sk_lr.coef_[0])
print("Scikit-learn intercept:", sk_lr.intercept_)

print("\nMyLR slope:", lr.m)
print("MyLR intercept:", lr.b)

print("\nSlope difference:", abs(sk_lr.coef_[0] - lr.m))
print("Intercept difference:", abs(sk_lr.intercept_ - lr.b))
```
